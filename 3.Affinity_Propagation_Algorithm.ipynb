```python
import pandas as pd
import numpy as np
from sklearn.cluster import AffinityPropagation
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
dataset = pd.read_csv("Mall_Customers.csv")

# Preprocess the data: Extract features for clustering
X = dataset[['Annual Income (k$)', 'Spending Score (1-100)']].values

# Apply Affinity Propagation clustering
affinity_propagation = AffinityPropagation()
labels = affinity_propagation.fit_predict(X)

# Number of clusters
n_clusters = len(np.unique(labels))
print("Number of clusters:", n_clusters)

# Visualize the clustering result
plt.figure(figsize=(10, 9))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', data=supervised, hue=labels, palette='Set2', legend='brief')
plt.title('Affinity Propagation Clustering of Mall Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.show()
```
