```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Step 1: Load the dataset
dataset = pd.read_csv("Mall_Customers.csv")

# Step 2: Preprocess the data
# Assume we are clustering based on the 'Annual Income (k$)' and 'Spending Score (1-100)'
X = dataset[['Annual Income (k$)', 'Spending Score (1-100)']].values

# Step 3: Apply DBSCAN Clustering
# eps is the maximum distance between two samples for one to be considered as in the neighborhood of the other.
# min_samples is the number of samples (or total weight) in a neighborhood for a point to be considered as a core point.
dbscan = DBSCAN(eps=11, min_samples=5)
labels = dbscan.fit_predict(X)

# Add cluster labels to the original DataFrame
dataset['Cluster'] = labels

# Step 4: Print the cluster labels
print("Cluster labels:\n", labels)

# Step 5: Visualize the clustering result
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataset, x=X[:, 0], y=X[:, 1], c=labels, hue='Cluster', palette='viridis', marker='o')
plt.title('DBSCAN Clustering of Mall Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
